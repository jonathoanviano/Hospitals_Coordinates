Starting from the hospitales2.csv file, generate an output CSV file that contains the following fields in this order:

    latitude
    longitude
    name
    label

With the corresponding data extracted from the original CSV, the "name" field should correspond to the hospital's address, and the "label" field should contain the hospital's name.'''

```python
import pandas as pd
```

```python
df = pd.read_csv('hospitales2.csv', sep = ',')  
df.head()
```

##  ETL or "Data Wrangling". 
Link:<br>
[Replace values in a DataFrame](https://stackoverflow.com/questions/37593550/replace-method-not-working-on-pandas-dataframe)<br><br>

The "POINT" and parentheses in the WKT column (coordinates) should be removed, leaving the latitude and longitude values in the same column.

```python
df["WKT"]= df.WKT.str.replace('POINT', '')
df["WKT"]= df.WKT.str.replace('(', '')
df["WKT"]= df.WKT.str.replace(')', '')
df.head()
```

## Renombre de columnas<br>
[Rename columns in Pandas](https://www.statology.org/pandas-rename-columns/)<br><br>
We rename 'WKT' as 'Coordenadas'

```python
df.rename(columns = {'WKT':'Coordenadas'}, inplace = True)
df.head(2)
```

First, we ensure that there are no blank spaces in the 'Coordenadas' column, and then we split the values in that column into 'Latitude' and 'Longitude', knowing that there is a space (' ') between both values.

```python
df['Coordenadas'] = df['Coordenadas'].str.lstrip()
```

```python
df[['Latitude', 'Longitude']] = df['Coordenadas'].str.split(' ', expand=True)
df.head()
```

Now we will rearrange the data.
First, we will delete the 'Coordenadas' column.
Then, we will place the 'Latitude' and 'Longitude' columns at the beginning of the DataFrame, simply for organizational purposes.

```python
df = df.drop('Coordenadas', axis=1)
df.head()
```

First, visualize the list of columns. Then, copy and paste them, and assign them the desired order.

```python
columnas = list(df.columns.values)
columnas
```

```python
# Defining new order

df = df[['Latitude',
    'Longitude',
    'ID',
 'NOMBRE',
 'NOM_MAP',
 'OBJETO',
 'CALLE',
 'ALTURA',
 'DOM_NORMA',
 'DOM_GEO',
 'TELEFONO',
 'GUARDIA',
 'FAX',
 'WEB',
 'COD_POSTAL',
 'TIPO',
 'TIPO_ESPEC',
 'MOD_AT_1',
 'MOD_AT_2',
 'DEPEND_ADM',
 'DIRECTOR']]
```

```python
# New order
df.head()
```

Now we generate a new DataFrame with the data required for the exercise.

```python
df_data = df[['Latitude',
    'Longitude',
    'NOMBRE',
    'DOM_NORMA']]
```

```python
df_data.rename(columns = {'NOMBRE':'Label','DOM_NORMA':'Name'}, inplace = True)
```

```python
'''I changed the order because when I printed it on https://www.gpsvisualizer.com/map?output_home, 
it placed the addresses in the middle of the ocean.'''
df_data = df_data.rename(columns={'Longitude':'Latitude',
                                   'Latitude':'Longitude'})

df_data.columns
```

```python
df_data
```

We need to check that the coordinate data is of type 'float'.

```python
print(type(df_data['Latitude'][0]))
print(type(df_data['Longitude'][0]))
```

Both are 'str' type, so we need to convert them to 'float'.

```python
df_data['Latitude'] = pd.to_numeric(df_data['Latitude'], downcast='float')
df_data['Longitude'] = pd.to_numeric(df_data['Longitude'], downcast='float')
print(type(df_data['Latitude'][0]))
print(type(df_data['Longitude'][0]))
```

Finally, we proceed to make a .CSV copy with the required data.

```python
# It is very important to include index=False when exporting to CSV, 
# otherwise it will include a comma and an index in the output.

df_data.to_csv('Data_Coordenadas.csv', index = False)
```
